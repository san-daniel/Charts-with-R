```r
# install.packages("fmsb", type = "source")
library(fmsb)
```

```r
pkm <- read.csv("https://raw.githubusercontent.com/san-daniel/Charts-with-R/master/pkm.csv", row.name = 1)
head(pkm)
colnames(pkm)
```

```r
# Graficas de Radar
```

```r
pkm.fav <- c("Greninja", "Metagross", "Pikachu")
```

```r
tmp <- pkm[pkm.fav, c('hp','attack','defense','sp_attack','sp_defense','speed')]
tmp
```

```r
tmp <- rbind(rep(255,6),rep(0,6),tmp)
tmp
```

```r
radarchart(tmp)
```

```r
colors_border=c( rgb(0.2,0.5,0.5,0.9), rgb(0.8,0.2,0.5,0.9) , rgb(0.7,0.5,0.1,0.9) )
colors_in=c( rgb(0.2,0.5,0.5,0.4), rgb(0.8,0.2,0.5,0.4) , rgb(0.7,0.5,0.1,0.4) )
```

```r
radarchart( tmp  , axistype=1 , 
    #custom polygon
    pcol=colors_border , pfcol=colors_in , plwd=4 , plty=1,
    #custom the grid
    cglcol="grey", cglty=1, axislabcol="grey", caxislabels=seq(0,255,5), cglwd=0.8,
    #custom labels
    vlcex=0.8 
    )

# Add a legend
legend(x=0.7, y=1, legend = pkm.fav, bty = "n", pch=20 , col=colors_in , text.col = "grey", cex=1.2, pt.cex=3)
```
